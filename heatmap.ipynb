```python
flights_long = sns.load_dataset("flights")
flights = (
    flights_long
    .pivot(index="month", columns="year", values="passengers")
)
```

```python
flights
```

```python
sns.heatmap(flights,annot=True,fmt="d",linewidths=1)
```

```python
sns.heatmap(flights,annot=True,fmt="d",linewidths=1, cbar=False)
```
